## Function 1: Synonym()

```python
#creating a function which takes a word and prints its synonym in synonym array
def synonym(amen):
  synonyms = []
  for syn in wn.synsets(amen):
    for l in syn.lemmas():
      synonyms.append(l.name())
  return synonyms
  #print(set(synonyms))
```

```python
# function that reads an array and output the sorted array according to frequency of unique values 
def arrangement(test):
# sorting on bais of frequency of elements
  result = sorted(test, key = test.count, reverse = True)
 #arrange according to unique characters 
  used = set()
  unique = [x for x in result if x not in used and (used.add(x) or True)]
  #print(unique)
  return(unique)
  #print(Counter(test)) 
 # print("Unique:      ", unique)
  #print("Size Unique: ", len(unique))
```

### Function: dictValue()

```python
# Create a dictionary of unique words, ordered according to the frequency
def dictValue(unique):
    valueSynm = []
    keys = [unique[0]]
    values = [synonym(unique[0])]
    diction = dict(zip(keys, values)) 
    valueSynm.extend(synonym(unique[0]))
    for x in range(1,len(unique)):
      if(unique[x] in valueSynm):
        continue
      else:
        valueSynm.extend(synonym(unique[x])) #create list/array of values
        diction[unique[x]] = synonym(unique[x]) # adding a dicitonary 
    print('Dictionary: ', diction)
    print('Dictionary Size:', len(diction))
    print('Dictionary Keys:', diction.keys())
    print('Dictionary Values:', valueSynm)
    print('Dictionary Values Size', len(valueSynm))
    #return(diction)
    #return (tempo)
```
